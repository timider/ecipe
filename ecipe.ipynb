```python
import pandas as pd
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQV0b_1tr9ZOwf_runoQMdEhFgePUXm5DfiAKlARhx_H8eR3XuDxGlCl-FYtTczaRuZXM1z5NnWgiu-/pub?output=csv'
df = pd.read_csv(url)
df.head()
```

```python
df = df.melt(id_vars=['Series Name', 'Country Name'], var_name='Year', value_name='Value')
df['Year'] = df['Year'].str.extract('(\d{4})')
df
```

```python
df.sort_values(by=['Country Name', 'Year'], inplace=True)
df
```

```python
df = df.pivot(index=['Country Name', 'Year'], columns='Series Name', values='Value')
df.columns.name = "Statistic"
df
```

```python
for col in df.columns:
  if col != 'Country Name' and col != 'Year' and df[col].dtype == 'object':
    df[col] = df[col].str.replace(',', '.').astype(float)
```

```python
df['All exports'] = df['Service exports (BoP, current US$)'] + df['Goods exports (BoP, current US$)']
df['Share of services exports'] = df['Service exports (BoP, current US$)'] / df['All exports']
df['Share of goods exports'] = df['Goods exports (BoP, current US$)'] / df['All exports']
df
```

```python
df = df.reset_index()
```

```python
import matplotlib as plt
import matplotlib.pyplot as plt
import numpy as np
```

```python
df['Country Name'].replace({'Brazil':'BRA'}, inplace=True) 
df['Country Name'].replace({'Colombia':'COL'}, inplace=True) 
df['Country Name'].replace({'Peru':'PER'}, inplace=True) 
df['Country Name'].replace({'Thailand':'THA'}, inplace=True) 
df['Country Name'].replace({'Chile':'CHI'}, inplace=True) 
df['Country Name'].replace({'Ecuador':'ECU'}, inplace=True) 
df['Country Name'].replace({'South Africa':'RSA'}, inplace=True) 
```

b) 

```python
df_2019 = df[df['Year'] == "2019"]
x = np.arange(len(df_2019))
plt.bar(x-0.2, df_2019['Share of services exports'], width = 0.4, label = 'Share of services exports', color = 'blue')
plt.bar(x+0.2, df_2019['Share of goods exports'], width = 0.4, label = 'Share of goods exports', color = 'red')
plt.xlabel('Country Name')
plt.ylabel('Share of Exports')
plt.title('Share of Services and Goods Exports in 2019')
plt.xticks(x, df_2019['Country Name']) 
plt.legend()
plt.show()
```

Out of the selected countries, Thailand's export basket is most reliant on services exports.

```python
df['Commercial services exports index 2005=1'] = df.groupby('Country Name')['Commercial service exports (current US$)'].transform(lambda x: x/x.iloc[0])
df
```

```python
df['Commercial services exprorts total growth rate (%)'] = df.groupby('Country Name')['Commercial services exports index 2005=1'].transform(lambda x: 100*((x/x.iloc[0]) - 1))
df
```

```python
df['Year'] = pd.to_numeric(df['Year'])
```

```python
df['Commercial services exports CAGR'] = df.groupby('Country Name')['Commercial services exports index 2005=1'].transform(lambda x: (x.astype(float)/x.iloc[0].astype(float))**(1/(df['Year']-2005)))
df
```

What is the relationship between the value index and the growth rate?



```python
df_col = df[df['Country Name'] == 'COL']
plt.plot(df_col['Year'], df_col['Commercial services exports index 2005=1'], marker='o')
plt.plot(df_col['Year'], df_col['Commercial services exports CAGR'], marker='^')
plt.xlabel('Years')
plt.ylabel('Value')
plt.title("Commercial services exports index and CAGR for Colombia")
plt.legend()
plt.show()
```

Value index and goods exports in total exports for 2019 for Colombia and peer countries.
